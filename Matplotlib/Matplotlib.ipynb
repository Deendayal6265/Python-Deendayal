```python
# plot function x^3
import numpy as np 
x = np.arange(0,5,0.1)
y = x**3
z = x**2
#plt.scatter(x,y)
#plt.plot(x,y,color = "red", marker = "p")
plt.plot(x,z,c="y",linewidth =3 ,label ="x^2")
plt.plot(x,y,color = "black" ,linewidth = 2,label = "x,y")

plt.axis([0,10,0,100])
plt.legend()
plt.text(4,60,"4,60")
plt.title("X^3 line graph ")
plt.grid()
plt.xlabel("x lebel",size=25,color="blue")
plt.ylabel("y lebel ",size=25,color="red")
plt.show()
```

#** Bubble Chart **

```python
import matplotlib.pyplot as plt 
```

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12, 13, 14 ,17, 19, 20]
population = [100, 120, 180 ,250,300 ,370]

```

```python
plt.scatter(year,salary,s=population,c=year,edgecolor="red",alpha=1) # alpha is increase or dec. the intesity of bubbles . )

plt.title("Bubble Chart ")
plt.xlabel("Year", size=25)
plt.ylabel("salary in thousands ",c="red",size=15)
plt.show()
```

```python
plt.scatter(year,salary,s=population,c=year,edgecolor="red",alpha=0.3,marker='^') # alpha is increase or dec. the intesity of bubbles . )

plt.title("Bubble Chart ")
plt.grid()
plt.text(2014,14.50,"HEllo guys")
plt.xlabel("Year", size=20,c='blue')
plt.ylabel("salary in thousands ",c="red",size=15)
plt.show()
```

# Pie Chart 

```python
number = [3,4,6,2]
labels = ["A","B","C","D"]

colors = ["red","blue","yellow","purple"]

explode = [0.1,0,0,0]  # it cut the pie and show like a cuttign pizza . here 0.1 is the gap between the real pie and cutting pie . 

plt.pie(number,colors=colors,labels=labels,explode=explode,autopct="%.2f", counterclock=False,startangle=120) #pie
#counter clock for pie in clock wiesed .

plt.title("Pie chart ",size=25,color="blue")

plt.axis("equal")   # it make our pie fully circle. 
plt.show()
```

# Bar Graph . 

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12, 13, 14 ,17, 19, 20]
population = [100, 120, 180 ,250,300 ,370]

plt.bar(year,population,edgecolor="black",color="purple",alpha=0.8,width=0.8)
plt.xticks(rotation=90)  # it rotate the x axis labels .
plt.title("Bar Graph ",color="red",size=25)
plt.xlabel("year",size=25)
plt.ylabel("population",size=20)
plt.grid()
plt.show()
```
